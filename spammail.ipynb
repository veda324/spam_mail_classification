```python
import numpy as np
import matplotlib.pyplot as plt
w = 0.8  
b = -3   
def relu(z):
    return np.maximum(0, z)
x_vals = np.linspace(0, 10, 100)   # spam score from 0 to 10
z_vals = w * x_vals + b            # weighted sum
y_vals = relu(z_vals)              # apply activation

plt.figure(figsize=(6,4))
plt.plot(x_vals, y_vals)
plt.xlabel("Spam Score (input)")
plt.ylabel("Neuron Output (ReLU)")
plt.title("Neuron Output vs Input (ReLU Activation)")
plt.grid(True)
plt.tight_layout()
plt.show()

spam_keywords = 6  

z = w * spam_keywords + b
output = relu(z)

print("Weighted sum (z):", z)
print("ReLU Output:", output)

# Decide if email is spam
if output > 0:
    print("Final Result: Email is SPAM")
else:
    print("Final Result: Email is NOT SPAM")
```
