```python
import numpy as np
x = np.array([1.0, 0.5, -1.2])
y_true = np.array([1.5, -0.5, 2.0, 0.0, 1.0, -1.0, 0.3, 2.5, -0.7, 1.2])
W = np.random.randn(3, 10)
b = np.zeros(10)
y_pred = np.dot(x, W) + b             # predicted output (10 values)
delta = 1.0
loss = 0.0

for i in range(10):
    error = y_true[i] - y_pred[i]

    if abs(error) <= delta:
        loss += 0.5 * error ** 2
    else:
        loss += delta * (abs(error) - 0.5 * delta)
print("Predicted values:", y_pred)
print("True values     :", y_true)
print("Huber Loss      :", loss)
```
