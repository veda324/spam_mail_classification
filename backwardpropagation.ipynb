```python
import math

# -------- Activation functions --------
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def sigmoid_derivative(y):
    return y * (1 - y)

# -------- Inputs --------
x1, x2 = 0.35, 0.7
target = 1
lr = 0.1   # learning rate

# -------- Initial weights --------
w11, w12 = 0.2, 0.3
w21, w22 = 0.2, 0.3
w13, w23 = 0.3, 0.9

# -------- Forward Pass --------
h1_in = x1*w11 + x2*w21
h1 = sigmoid(h1_in)

h2_in = x1*w12 + x2*w22
h2 = sigmoid(h2_in)

o_in = h1*w13 + h2*w23
output = sigmoid(o_in)

# -------- Loss (MSE) --------
loss = 0.5 * (target - output) ** 2
print("Output:", round(output, 4))
print("Loss:", round(loss, 4))

# -------- Backward Pass --------
# Output layer error
error_out = (output - target) * sigmoid_derivative(output)

# Hidden layer errors
error_h1 = error_out * w13 * sigmoid_derivative(h1)
error_h2 = error_out * w23 * sigmoid_derivative(h2)

# -------- Update weights --------
w13 -= lr * error_out * h1
w23 -= lr * error_out * h2

w11 -= lr * error_h1 * x1
w21 -= lr * error_h1 * x2

w12 -= lr * error_h2 * x1
w22 -= lr * error_h2 * x2

print("\nUpdated Weights:")
print("w11:", round(w11,4), "w21:", round(w21,4))
print("w12:", round(w12,4), "w22:", round(w22,4))
print("w13:", round(w13,4), "w23:", round(w23,4))
```
